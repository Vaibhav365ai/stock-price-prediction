```python
import numpy as np

def value_plot(df, y, sort_ascending=False, figscale=1):
  from matplotlib import pyplot as plt
  figsize = (8 * figscale, 4 * figscale)
  if sort_ascending:
    df = df.sort_values(y).reset_index(drop=True),
    ax = plt.subplots(figsize=figsize)
  df[y].plot(kind='line')
  plt.title(y)


chart = value_plot(df, ['Open'], *{})
chart

chart1 = value_plot(df, ['Adj Close'], *{})
chart1
```

```python
df.replace('Date',1,inplace=True)
```

```python
x=df.iloc[0:,[1,2,3]]
y=df.iloc[:,[4]]
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=100,random_state=0)
```

```python
regressor.fit(xtrain,ytrain)
```

```python
ytrain_pred = regressor.predict(xtrain)
```

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
ytrain_pred = regressor.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))
r2 = r2_score(ytrain, ytrain_pred)
```

```python
print("The model performance for training set")
print("--------------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

```python
ytest_pred = regressor.predict(xtest)
mse = mean_squared_error(ytest, ytest_pred)
rmse = (np.sqrt(mean_squared_error(ytest, ytest_pred)))
r2 = r2_score(ytest, ytest_pred)
```

```python
print("The model performance for testing set")
print("--------------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
